```python
import urllib.request
url = "http://mlg.ucd.ie/modules/python/EVdata"
response = urllib.request.urlopen(url)
html = response.read().decode()

lines = html.strip().split('\n')
for l in lines:
    print(l)
```

```python
import bs4
from urllib.request import urlopen, urlretrieve, quote

parser = bs4.BeautifulSoup(html,'html.parser')
for link in parser.find_all('a'):
    full_link = 'http://mlg.ucd.ie/modules/python/EVdata/'+ link.get('href')
    #print(full_link) # EuroVelo Route List
    response2 = urllib.request.urlopen(full_link)
    html2 = response2.read().decode()
    inner_parser = bs4.BeautifulSoup(html2,'html.parser')
    for l in inner_parser.find_all('a'):       
        full_link2 = 'http://mlg.ucd.ie/modules/python/EVdata/'+l.get('href')
        #print(full_link2) # GPX file
        #response3 = urllib.request.urlopen(full_link2)
        #xml = response3.read().decode()
        filename = l.get('href').rsplit('/', 1)[-1]
        print("Downloading %s to %s..." % (full_link2, filename) )
        urlretrieve(full_link2, filename)
        print("Done.")
```
